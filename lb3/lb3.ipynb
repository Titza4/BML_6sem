```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import tree
from sklearn import model_selection
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn import metrics
```

```python
voice_data = pd.read_csv('voice_gender.csv')
voice_data.head()

# Define features and target
X = voice_data.drop('label', axis=1)
y = voice_data['label']
```

```python
voice_data.isnull().sum().sum()
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
print('Train shape: {}'.format(X_train.shape))
print('Test shape: {}'.format(X_test.shape))
```

```python
# Создание и обучение дерева решений максимальной глубины 1 с критерием 'entropy'
dt_clf = tree.DecisionTreeClassifier(max_depth=1, criterion='entropy', random_state=42)
dt_clf.fit(X_train, y_train)

# Визуализация дерева решений
plt.figure(figsize=(14, 10), dpi=100)
tree.plot_tree(dt_clf, feature_names=X.columns, class_names=dt_clf.classes_, filled=True, rounded=True, fontsize=16)
plt.title('Decision Tree (max_depth=1, criterion=entropy)')
plt.show()
```

```python

feature = X.columns[dt_clf.tree_.feature[0]]
print("Корневое правило построено по признаку:", feature)
```

```python
threshold = dt_clf.tree_.threshold[0]
print("Оптимальное пороговое значение:", round(threshold, 3))
```

```python
left_condition = X_train[feature] <= threshold
percent = left_condition.mean() * 100
print("Процент наблюдений, удовлетворяющих условию:", round(percent, 1))
```

```python
y_pred = dt_clf.predict(X_test)
acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy на тестовой выборке:", round(acc, 3))
```

Задание 2

```python
# Обучим дерево глубины 2
dt2 = tree.DecisionTreeClassifier(max_depth=2, criterion='entropy', random_state=42)
dt2.fit(X_train, y_train)

# Получим индексы признаков, используемых в дереве
used_features = set(dt2.tree_.feature)
used_features.discard(-2)  # -2 означает "нет признака" (лист)
used_feature_names = [X.columns[i] for i in used_features]
print("Используемые признаки:", used_feature_names)
```

```python
# Визуализация дерева
plt.figure(figsize=(14, 10), dpi=100)
tree.plot_tree(dt2, feature_names=X.columns, class_names=dt2.classes_, filled=True, rounded=True, fontsize=16)
plt.title('Decision Tree (max_depth=2, criterion=entropy)')
plt.show()

# Подсчёт листьев с предсказанием "female"
leaf_indices = np.where(dt2.tree_.children_left == -1)[0]
female_leaves = sum(dt2.classes_[np.argmax(dt2.tree_.value[i])] == 'female' for i in leaf_indices)
print("Листьев с классом female:", female_leaves)
```

```python
y_pred2 = dt2.predict(X_test)
acc2 = metrics.accuracy_score(y_test, y_pred2)
print("Accuracy на тестовой выборке:", round(acc2, 3))
```

Задание 3

```python
# Дерево решений без ограничения глубины
full_tree = tree.DecisionTreeClassifier(criterion='entropy', random_state=0)
full_tree.fit(X_train, y_train)

# 1. Глубина дерева
print("Глубина дерева:", full_tree.get_depth())

# 2. Количество листьев
print("Количество листьев:", full_tree.get_n_leaves())

# 3. Accuracy на обучающей и тестовой выборках
train_pred = full_tree.predict(X_train)
test_pred = full_tree.predict(X_test)
train_acc = metrics.accuracy_score(y_train, train_pred)
test_acc = metrics.accuracy_score(y_test, test_pred)
print("Accuracy на обучающей выборке:", round(train_acc, 3))
print("Accuracy на тестовой выборке:", round(test_acc, 3))


```

Задание 4

```python
# Задаём сетку параметров
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [4, 5, 6, 7, 8, 9, 10],
    'min_samples_split': [3, 4, 5, 10]
}
```

```python
# Задаём метод кросс-валидации
cv = StratifiedKFold(n_splits=5)
```

```python
# Инициализируем GridSearchCV
grid_search = GridSearchCV(
    estimator=tree.DecisionTreeClassifier(random_state=0),
    param_grid=param_grid,
    scoring='accuracy',
    cv=cv,
    n_jobs=-1
)

# Обучаем GridSearchCV на обучающей выборке
grid_search.fit(X_train, y_train)

# 1. Лучший критерий информативности
print("Лучший критерий информативности:", grid_search.best_params_['criterion'])

# 2. Оптимальная максимальная глубина
print("Оптимальная max_depth:", grid_search.best_params_['max_depth'])

# 3. Оптимальное min_samples_split
print("Оптимальное min_samples_split:", grid_search.best_params_['min_samples_split'])

# 4. Оценка на обучающей и тестовой выборках
best_model = grid_search.best_estimator_
train_acc = metrics.accuracy_score(y_train, best_model.predict(X_train))
test_acc = metrics.accuracy_score(y_test, best_model.predict(X_test))
print("Accuracy на обучающей выборке:", round(train_acc, 3))
print("Accuracy на тестовой выборке:", round(test_acc, 3))

```

```python
# ВАЖНОСТЬ ПРИЗНАКОВ
importances = best_model.feature_importances_
feature_names = X.columns

# Визуализация важности признаков
plt.figure(figsize=(12, 6))
indices = np.argsort(importances)[::-1]
plt.bar(range(len(importances)), importances[indices], align='center')
plt.xticks(range(len(importances)), feature_names[indices], rotation=90)
plt.title('Feature Importances (Decision Tree)')
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.tight_layout()
plt.show()

# Топ-3 наиболее важных фактора
for i in range(3):
    print(f"Топ-{i+1}: {feature_names[indices[i]]} (важность: {importances[indices[i]]:.3f})")
```
